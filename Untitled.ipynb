```python
import folium
# [위도, 경도]
center = [35.8242238, 127.1479532]
map = folium.Map(location = center, zoom_start=11.5)
map
```

```python
import pandas as pd
park = pd.read_csv('C:/Users/hyunj/Desktop/전주공모전자료/전국주차장/전라북도_전주시_주차장정보_20220311_1647220760617_29835.csv', encoding='CP949')
park.head()
park.info()
```

```python
import json

geo_data = json.load(open('layer.json', encoding='utf-8'))

folium.Choropleth(
    geo_data=geo_data,
    fill_color='white'
).add_to(map)
```

```python
for n in park.index:
    sublat = park.loc[n, '위도']
    sublong = park.loc[n, '경도']
    number = int(park.loc[n, '주차구획수'])
    folium.Circle(location=[sublat, sublong], radius = number*0.5, color='red', fill_color='red').add_to(map)
```
